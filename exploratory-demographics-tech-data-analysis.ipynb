**EXPLORATORY DATA ANALYSIS**

```python
sns.set(rc={'figure.figsize':(12,5)})
plt.xticks(fontsize=15,rotation='vertical')
p=sns.countplot(df['Age'],hue_order=df.groupby('Age'))
p.axes.set_title("age proportions",fontsize=30)
plt.show()
```

```python
plt.figure(figsize = (16,8))
sns.countplot(data=df, y="Country", order=df['Country'].value_counts().index[:10])
plt.title('Top 10 Country', size=18)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(12,2)})
plt.xticks(fontsize=15,rotation='vertical')
p=sns.countplot(df['Gender'],hue_order=df.groupby('Gender'))
p.axes.set_title("gender distribution",fontsize=30)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(12,5)})
plt.xticks(fontsize=15,rotation='vertical')
p=sns.countplot(df['DatabaseWorkedWith'],hue_order=df.groupby('Gender'))
p.axes.set_title("Database Worked type",fontsize=30)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(12,5)})
plt.xticks(fontsize=15,rotation='vertical')
p=sns.countplot(df['SocialMedia'],hue_order=df.groupby('Gender'))
p.axes.set_title("Most Used social media",fontsize=30)
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.xticks(fontsize=15,rotation='vertical')
ax = sns.countplot(x="SocialMedia", hue="Student", data=df)
plt.title("SocialMedia used by studemt",fontsize=30)
plt.show()
```

```python
plt.figure(figsize = (16,8))
sns.countplot(data=df, y="DevEnviron", order=df['DevEnviron'].value_counts().index[:10])
plt.title('Top 10 Devloper Environ', size=18)
plt.show()
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import pandas as pd

comment_words = ''
stopwords = set(STOPWORDS)

# iterate through the csv file
for val in df.LanguageWorkedWith:

# typecaste each val to string
 val = str(val)

# split the value
 tokens = val.split()

# Converts each token into lowercase
 for i in range(len(tokens)):
  tokens[i] = tokens[i].lower()

 comment_words += " ".join(tokens)+" "

wordcloud = WordCloud(width = 800, height = 800,
background_color ='white',
stopwords = stopwords,
min_font_size = 10).generate(comment_words)

# plot the WordCloud image
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)

plt.show()
```
